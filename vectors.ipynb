```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')
ax.plot([0,x[0]], [0,y[0]], [0,z[0]])
ax.plot([0,x[1]], [0,y[1]], [0,z[1]])


ax.scatter(x[0], y[0], z[0], c='b', marker='o')
ax.scatter(x[1], y[1], z[1], c='r', marker='o')

plt.show()
```

```python
from math import sqrt, acos, degrees
 
x1 = x[0]
y1 = y[0]

 
x2 = x[1]
y2 = y[1]

 
def scalar(x1, y1, x2, y2):
    return x1*x2 + y1*y2
 
def module(x, y):
    return sqrt(x ** 2 + y ** 2)
 
cos = scalar(x1, y1, x2, y2)/(module(x1,y1)*module(x2,y2))
 
ang = acos(cos)
 
print(degrees(acos(cos)))
```
