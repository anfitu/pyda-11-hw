```python
import pandas as pd
```

```python
# Задача 1
data = pd.read_csv('C:\\Users\\Professional\\Desktop\\ratings.csv')
data_co = data.groupby('userId').count()[['rating']]
data100 = data_co[data_co.rating > 100]
```

```python
maxT = data.groupby('userId').max()[['timestamp']]
minT = data.groupby('userId').min()[['timestamp']]
rt = maxT.merge(minT, on='userId')
```

```python
def delta (row):
    return row['timestamp_x'] - row['timestamp_y']
rt['delta'] = rt.apply(delta, axis = 1)
```

```python
rt.delta
```

```python
result_table = data100.merge(rt, on = 'userId')
sum(result_table.delta) / len(result_table.delta)
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table = pd.concat([rzd, auto, air], axis = 0) #общая таблица со всеми данными
table.drop_duplicates(subset = 'client_id', keep = 'first', inplace = True) # уникальные клиенты
clients = table.client_id # датафрейм, состоящий из списка уникальных клиентов
cr = pd.merge(clients, rzd, on = 'client_id', how = 'left') # попытка присоединения первой из таблиц к списку клиентов
cra = pd.merge(cr, auto, on = 'client_id', how = 'left')
craa = pd.merge(cra, air, on = 'client_id', how = 'left')
craa
```

```python
res_tab = pd.merge(craa, client_base, on = 'client_id')
res_tab
```
