```python
data = pd.read_csv('C:\\Users\\Professional\\Desktop\\ratings.csv').head()
data
```

```python
def films_3 (row):
    if row.rating <= 4 and row.rating > 2:
        return 'Средний рейтинг'
    if row.rating <= 2:
        return 'Низкий рейтинг'
    if row.rating > 4:
        return 'Высокий рейтинг'
data['class'] = data.apply(films_3, axis = 1)
data.head()
```

```python
# Задача 2
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
cities = pd.read_csv('C:\\Users\\Professional\\Desktop\\keywords.csv')
def regions(keyword):
    for k, v in geo_data.items():
        for i in v:
            if i in keyword:
                return k
        return  'undefined'
cities['region'] = cities['keyword'].apply(regions)
cities.head(20)
```
